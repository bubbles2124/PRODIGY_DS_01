```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Create Dataframe
df = pd.read_csv("Population.csv")
```

```python
# Set style
sns.set(style="whitegrid")
```

```python
# Create a bar chart
plt.figure(figsize=(14, 8))
sns.barplot(x='2022 Population', y='Country/Territory', data=df, palette='viridis')

# Add title and labels
plt.title('Population by Country in 2022')
plt.xlabel('Population')
plt.ylabel('Country')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a histogram for World Population Percentage by Continent
plt.figure(figsize=(14, 8))
sns.histplot(data=df, x='World Population Percentage', hue='Continent', multiple='stack', palette='viridis', kde=True)

# Add title and labels
plt.title('World Population Percentage by Continent in 2022')
plt.xlabel('World Population Percentage')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```
